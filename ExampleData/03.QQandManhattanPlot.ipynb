```python
assoc_df=pd.read_csv("./test.assoc",delim_whitespace=True)
assoc_df
```

```python
import numpy as np
from itertools import cycle

CHR="CHR"
POS="BP"
P="P"
Assoc_df=assoc_df

Assoc_df[CHR]=Assoc_df[CHR].replace("X","23");Assoc_df[CHR]=Assoc_df[CHR].replace("Y","24")
Assoc_df[CHR]=Assoc_df[CHR].astype(int)

last_xpos = 0
x, y, c= [], [], []
colors=cycle(["#3B5488", "#53BBD5"])
xs_by_id = []
for seqid, group_data in Assoc_df.groupby(by=CHR, sort=False):  # keep the raw order of chromosome
    color = next(colors)
    for i, (site, p_value) in enumerate(zip(group_data[POS], group_data[P])):
        x.append(last_xpos + site)
        y.append(-np.log10(p_value))
        c.append(color)
    xs_by_id.append([seqid, last_xpos + (group_data[POS].iloc[0] + group_data[POS].iloc[-1]) / 2])
    last_xpos = x[-1]
        
fig, ax = plt.subplots(figsize=(15, 4))
plt.scatter(data=Assoc_df, x=x, y=y, s=2, color=c)
ax.set_xticks([v for c, v in xs_by_id], labels=[c for c, v in xs_by_id])
ax.set_xlim(x[0]-100000, x[-1]+100000)
ax.axhline(y=-np.log10(0.05/len(Assoc_df)), color="green")
ax.axhline(y=-np.log10(1/len(Assoc_df)), color="red")
```
